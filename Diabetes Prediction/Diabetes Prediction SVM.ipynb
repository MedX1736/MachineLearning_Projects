```python
#Import Depends 
import pandas as pd 
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 
from sklearn import svm 
from sklearn.metrics import accuracy_score
```

```python
#Data 
data = pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
print("The Data has: " + str(data.shape[0]) + " Rows" ) 
print("The Data has: " + str(data.shape[1]) + " Columns" ) 
print("Which are: ")
print(data.columns)
```

```python
data.describe()
```

```python
data['Outcome'].value_counts()
```

```python
data.groupby('Outcome').mean()
```

```python
# Separating Data and labels 
X = data.drop(columns='Outcome',axis = 1)
Y = data['Outcome']
```

```python
print('Data Shape : ',X.shape)
print('Lables Shape : ',Y.shape)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
```

```python
#Split data into training and test Data 
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8 , stratify=Y , random_state = 2 )
```

```python
print(X.shape,X_train.shape,X_test.shape)
print(Y.shape,Y_train.shape,Y_test.shape)
```

```python
#Training the model 
classifier  = svm.SVC(kernel = 'linear')
```

```python
 classifier.fit(X_train,Y_train)
```

```python
#Evaluating Model 
predictions =classifier.predict(X_test)
print("Accuracy : ", accuracy_score(predictions,Y_test))
```

```python
input_data = (1,89,76,34,37,31.2,0.192,23)
input_as_np = np.asarray(input_data)
reshaped_data = input_as_np.reshape(1,-1)
standardized_input = scaler.transform(reshaped_data)
```

```python
prediction = classifier.predict(standardized_input)
```

```python
if (prediction[0] == 1 ) : 
  print("This person has diabetes")
elif (prediction[0] == 0) : 
  print("This person doesn't have diabetes")
```

```python
# Logistic regression 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
predictions = model.predict(X_test)
print("Accuracy Score : ",accuracy_score(predictions,Y_test))
```
