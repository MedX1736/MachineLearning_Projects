```python
#Import Dependencies 
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn import svm 
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("parkinsons.csv")
```

```python
#Features and Traget 
X = df.drop(columns = ['status','name'],axis = 1)
Y= df['status']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=2,stratify=Y)
```

```python
print(X_train.shape,X_test.shape)
```

```python
#standardisation of data
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Training model
model = svm.SVC(kernel="linear")
```

```python
model.fit(X_train,Y_train)
```

```python
#Evaluating Model on Training Data
pred_train = model.predict(X_train)
print("Accuracy Score on Tranining Data : ", accuracy_score(Y_train,pred_train))
```

```python
#Evaluating Model on Training Data
pred_test = model.predict(X_test)
print("Accuracy Score on Tranining Data : ", accuracy_score(Y_test,pred_test))
```

```python
# Predictive System 
input_data = np.asarray((119.99200,157.30200,74.99700,0.00784,0.00007,0.00370,0.00554,0.01109,0.04374,0.42600,0.02182,0.03130,0.02971,0.06545,0.02211,21.03300,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654))

reshaped = input_data.reshape(1,-1)

scaled_input = scaler.transform(reshaped)

prediction = model.predict(scaled_input)
```
